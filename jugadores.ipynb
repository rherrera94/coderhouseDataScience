#Problema a resolver: ¿ a que provincias pertenecen los jugadores de la seleccion argentina?
##Pasos para resolver el problema:
- Encontrar los jugadores de la seleccion.
- encontrar los perfiles de los jugadores.
- Conocer la ciudad a la que pertenecen los jugadores.
- Conocer la provincia de la ciudad.

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
webjugador='https://es.wikipedia.org/wiki/Categor%C3%ADa:Futbolistas_de_Argentina_en_la_Copa_Mundial_de_2022'
responsejugador = requests.get(webjugador)
contentjugador = responsejugador.text
soupjugador = BeautifulSoup(contentjugador, 'lxml')
datosjugador = soupjugador.find_all('div', class_='mw-category mw-category-columns')
```

```python
ajugadores=datosjugador[0].find_all('a')
```

```python
urls=[]
nombres=[]
numero=0
for referencias in ajugadores:
  urls=np.append(urls, referencias.get('href'), axis=None)
  nombres=np.append(nombres, referencias.get('title'), axis=None)
```

```python
def funcion(jugador):
  web = f'https://es.wikipedia.org/{jugador}'
  response = requests.get(web)
  content = response.text
  soup = BeautifulSoup(content, 'lxml')
  datos = soup.find_all('div', class_='mw-parser-output')
  tablas=datos[0].find_all('table')
  datos=str(datos)
  tablas=tablas[0]
  tablas=str(tablas)
  datos=datos.split(tablas)
  datos=','.join(datos)
  datos=BeautifulSoup(datos,'lxml')
  aux1=datos.find_all('p')
  aux=aux1[1].find_all('a')
  if(len(aux)!=0):
    return aux[0]
  return "error"

ciudadjugador=[]
linkciudad=[]

for jugador in urls:
    jugadorn=funcion(jugador)
    print(jugador)
    if(jugadorn!="error"):
      ciudadjugador=np.append(ciudadjugador, jugadorn.get_text(), axis=None)
      linkciudad=np.append(linkciudad, jugadorn.get('href'), axis=None)
    if(jugadorn=="error"):
      ciudadjugador=np.append(ciudadjugador, "Error", axis=None)
      linkciudad=np.append(linkciudad, "Error", axis=None)
```

```python
dict_MUNDIAL = {'Jugador': nombres, 'CiudadNacimiento': ciudadjugador,'linkciudad':linkciudad}

df_mundial = pd.DataFrame(dict_MUNDIAL)

df_mundial['año'] = '2022'
```

```python
df_mundial
```

```python
def funcion_provincia(jugador):
  web = f'https://es.wikipedia.org{jugador}'
  response = requests.get(web)
  content = response.text
  soup = BeautifulSoup(content, 'lxml')
  datos = soup.find_all('div', class_='mw-parser-output')
  tot=len(datos)-1
  tablas=datos[tot].select('p > a')
  encontrado=False
  i=0
  while (encontrado==False and i<len(tablas)):
    if("Provincia de" in str(tablas[i])):
      encontrado=True
      objetivo=i
    i=i+1
  if(encontrado!=False):
    return tablas[objetivo].get('title')
  return "Error"
provjugador=[]
for i in range(len(linkciudad)):
  if (linkciudad[i]!='/wiki/Ciudad_de_Buenos_Aires'):
    prov=funcion_provincia(linkciudad[i])
    if(prov!="error"):
      provjugador=np.append(provjugador, str(prov), axis=None)
    if(prov=="error"):
      provjugador=np.append(provjugador, "Error", axis=None)
  else:
    provjugador=np.append(provjugador, "Ciudad de Buenos Aires", axis=None)
```

```python
df_mundial['provinciajug'] = provjugador
```
