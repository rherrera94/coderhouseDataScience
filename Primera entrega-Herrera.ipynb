Con el fin de poder realizar la matriz de correlación y lograr considerar cuales son los items con los cuales me quedaré para el análisis se hace necesaria la codificación de las variables categóricas

```python
dict_gen={'Masculino':0,'Femenino':1}
aerolinea['genero'] = aerolinea.genero.replace(dict_gen)
dict_viaje={'vacaciones':0,'Viaje de negocios':1}
aerolinea['tipo_viaje'] = aerolinea.tipo_viaje.replace(dict_viaje)
dict_clase={'Eco':0,'Eco Plus':1,'Business':2}
aerolinea['clase'] = aerolinea.clase.replace(dict_clase)
dict_tipo={'disloyal Customer':0,'Loyal Customer':1}
aerolinea['tipo_cliente'] = aerolinea.tipo_cliente.replace(dict_tipo)
dict_satisfaccionv2={'neutral o no satisfecho':0,'satisfecho':1}
aerolinea['satisfaccion'] = aerolinea.satisfaccion.replace(dict_satisfaccionv2)
```

En este caso se dio mayor importancia a aquellas columnas que tienen una cierta correlación que me hacen intuir que ayudaran a realizar un mejor análisis. Esto es asi ya que se observa, de los análisis previos, que aquellas personas que valoran de una forma más positiva al servicio prestado seran aquellos que mas satisfechos están y aquellos que valoran de manera menos positiva son aquellos que estan disconformes con el servicio. 

Por lo expuesto es se toma una variable target (que posteriormente pasara a ser la Y de nuestro análisis), la satisfacción.

```python
#Se genera un nuevo dataframe con los features relevantes y la variable target
analisis=pd.DataFrame()
analisis['tipo_cliente']=aerolinea['tipo_cliente']
analisis['edad']=aerolinea['edad']
analisis['tipo_viaje']=aerolinea['tipo_viaje']
analisis['clase']=aerolinea['clase']
analisis['distancia_vuelo']=aerolinea['distancia_vuelo']
analisis['wifi_abordo']=aerolinea['wifi_abordo']
analisis['facilita_compra']=aerolinea['facilita_compra']
analisis['comida_bebida']=aerolinea['comida_bebida']
analisis['checkin_online']=aerolinea['checkin_online']
analisis['comodidad_asiento']=aerolinea['comodidad_asiento']
analisis['entretenimiento_vuelo']=aerolinea['entretenimiento_vuelo']
analisis['servicio_abordo']=aerolinea['servicio_abordo']
analisis['espacio_piernas']=aerolinea['espacio_piernas']
analisis['equipaje']=aerolinea['equipaje']
analisis['servicio_checkin']=aerolinea['servicio_checkin']
analisis['inflight_service']=aerolinea['inflight_service']
analisis['limpieza']=aerolinea['limpieza']
analisis['satisfaccion']=aerolinea['satisfaccion']
analisis.info()
```

```python
#Se selecciona la variable target
x=analisis.drop(['satisfaccion'],axis=1)
y=analisis['satisfaccion']
```

```python
#Imports necesarios
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.naive_bayes import GaussianNB
```

Se separará entre grupos de entrenamiento y testeo (ambos x e y), la proporción que se elige es 70% para train y 30% para test. A su vez, para lograr mantener la proporción existente en las respuestas de satisfacción, tanto en train como en test, lo que se hace es usar stratify que me permite lograr esto. También se utiliza el random_state con la finalidad de producir una semilla que permita que los resultados se puedan replicar.

```python

rng = np.random.RandomState(23)
xentrenamiento, xtesteo,yentrenamiento,ytesteo = train_test_split(x,y, test_size=0.30, random_state=rng,stratify=y) 
```

#**Algoritmos de clasificación**

Teniendo en cuenta que para el análisis a realizar se cuenta con la variable target que es la respuesta a lo que necesito saber, en este caso la satisfaccion  o no del cliente, y que con la información historica que poseemos de las encuestas de satisfacción realizada intentaremos realizar predicciones sobre la satisfacción del cliente es que se elige la rama de algoritmos de aprendizaje supervisado. En éste caso elijo la rama de clasificación ya que necesito clasificar a los clientes entre satisfechos o no satisfechos, por lo tanto, éste será un problema de clasificación binaria. 

Por último, lo que se buscará al realizar un manejo de los parámetros de los algoritmos asi como también de lo realizado en la función train_test_split es que las métricas, de la seccion de test, de precisión y accuracy aumenten. Es de señalar que lo que muestro en ésta sección es producto de un refinamiento de los parámetros ya que se fue realizando pruebas con el fin de que los valores tanto de precisión como de accuracy aumenten y, a su vez, se busco también la disminución del overfitting que en un inicio tenian todos los modelos.
