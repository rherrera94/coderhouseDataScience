# Análisis de los estudiantes adultos de la enseñanza media técnico profesional

Se busca conocer la distribución, según el genero del alumno, dentro de las diferentes ramas de enseñanza

```python
#Estamos buscando agrupar por rama segun sexo la enseñanza técnico profesional de adultos, por lo tanto, primero haremos una diferenciación de la información general que necesitamos
adultotp_aux=rendimiento[(rendimiento["COD_ENSE2"]==8) & (rendimiento["COD_RAMA"].isin([400,500,600,700,800,900])) & (rendimiento["GEN_ALU"].isin([1,2])) ]
#Habiendo obtenido el universo de información que necesitamos, solo nos quedaremos con las columnas que queremos y ademas agruparemos segun la rama de estudio y el genero del alumno
criterio_adultotp={'COD_RAMA':'count'}
situacion_adultotp= adultotp_aux.groupby(['COD_RAMA','GEN_ALU']).agg(criterio_adultotp)
situacion_adultotp.columns=['CANTIDAD']
#convierto la serie en varias columnas nuevamente
situacion_adultotp=situacion_adultotp.reset_index()
situacion_adultotp
```

## Códigos de rama
| Código de rama  |   Rama Media Técnico Profesional  |
|---|---|
| 400   |   Comercial  |
| 500   |   Industrial  |
| 600   |   Técnica  |
| 700   |   Agrícola  |
| 800   |   Marítima  |


## Género alumno

| Código de genero  |   genero  |
|---|---|
| 1   |   Masculino  |
| 2   |   Femenino  |

```python
dict_codrama = {400:'Comercial', 500:'Industrial', 600:'Técnica', 700:'Agrícola', 800: 'Marítima'}
dict_genalu={1:'Masculino',2:'Femenino'}
situacion_adultotp['COD_RAMA'] = situacion_adultotp.COD_RAMA.replace(dict_codrama)
situacion_adultotp['GEN_ALU'] = situacion_adultotp.GEN_ALU.replace(dict_genalu)
fig_situacion_adultotp = px.bar(situacion_adultotp, x='COD_RAMA', y='CANTIDAD', color='GEN_ALU', barmode='relative',
             height=500,width=1000, text_auto=True,title="Educación adultos diferenciada según la rama y el genero del alumno")
fig_situacion_adultotp.show()
```

## Análisis gráfico

En el gráfico que antecede se puede ver una clara predominancia del género masculino dentro de la rama Industrial, mientras que en la rama Técnica se muestra una clara predominancia femenina. Otro punto a destacar es la muy baja cantidad de alumnos para la rama marítima con tan solo 14 alumnos. Por último se puede ver una predominancia femenina en la rama comercial.

# Análisis alumnos reprobados de los estudiantes adultos de la enseñanza media técnico profesional y Humanístico cientifico.

Se busca conocer la distribución de los reprobados, según el genero del alumno, dentro de las diferentes enseñanzas elegidas

```python
#Estamos buscando agrupar por rama segun sexo la enseñanza técnico profesional de adultos y la Educación Media Humanístico Científica Adultos, por lo tanto, primero haremos una diferenciación de la información general que necesitamos
adultotphc_aux=rendimiento[(rendimiento["COD_ENSE2"].isin([6,8])) &  (rendimiento["GEN_ALU"].isin([1,2])) &  (rendimiento["SIT_FIN"]=="R")]
#Habiendo obtenido el universo de información que necesitamos, solo nos quedaremos con las columnas que queremos y ademas agruparemos segun la rama de enseñanza y el genero del alumno
criterio_adultotphc={'COD_ENSE2':'count'}
situacion_adultotphc= adultotphc_aux.groupby(['COD_ENSE2','GEN_ALU']).agg(criterio_adultotphc)
situacion_adultotphc.columns=['CANTIDAD']
#convierto la serie en varias columnas nuevamente
situacion_adultotphc=situacion_adultotphc.reset_index()
situacion_adultotphc
```

## Códigos de enseñanza
| Código de enseñanza  |   Rama Media Técnico Profesional  |
|---|---|
| 6   |   Educación Media Humanístico Científica Adultos.  |
| 8   |   Educación Media Técnico Profesional y Artística, Adultos.  |

## Género alumno

| Código de genero  |   genero  |
|---|---|
| 1   |   Masculino  |
| 2   |   Femenino  |

```python
dict_codense={6:'Educación Media Humanístico Científica Adultos.',8:'Educación Media Técnico Profesional y Artística, Adultos.'}
situacion_adultotphc['COD_ENSE2'] = situacion_adultotphc.COD_ENSE2.replace(dict_codense)
dict_genalu={1:'Masculino',2:'Femenino'}
situacion_adultotphc['GEN_ALU'] = situacion_adultotphc.GEN_ALU.replace(dict_genalu)
fig_situacion_adultotphc = px.bar(situacion_adultotphc, x='COD_ENSE2', y='CANTIDAD', color='GEN_ALU', barmode='overlay',
             height=700,width=1000, text_auto=True,title="Reprobados en Educación adultos diferenciada según la enseñanza y el genero del alumno")
fig_situacion_adultotphc.show()
```

## Análisis del gráfico

En el gráfico que antecede se puede ver una predominancia, en ambos códigos de enseñanza, de reprobación masculina, destacandose la diferencia existente en la Educación Media Humanístico Científica Adultos. 

# Análisis de la situación final de los estudiantes con respecto a su asistencia a clase, según regiones

Se busca conocer la relación entre los alumnos reprobrados y sus inasistencias a clase y poder compararlos para ver si existe alguna marcada diferencia con respecto a los que mas asisten en promedio

```python
#agrupo la situacion final y las regiones para saber de los promovieron, reprobaron y fueron retirados el promedio de asistencia(si el espacio esta en blanco no hay informacion) 
criterio_situacion_final={'ASISTENCIA':'mean'}
situacion_final= rendimiento.groupby(['NOM_REG_RBD_A','SIT_FIN']).agg(criterio_situacion_final)
#convierto la serie en varias columnas nuevamente
situacion_final=situacion_final.reset_index()
situacion_final
```

## Códigos de situación
| Código de situación  |   Situación  |
|---|---|
| P   |   Promovido  |
| R   |   Reprobado  |
| Y   |   Retirado  |
| En blanco   |   Sin información  |

```python
dict_situ_final={'P':'Promovido','R':'Reprobado','Y':'Retirado'}
situacion_final['SIT_FIN'] = situacion_final.SIT_FIN.replace(dict_situ_final)
fig_situacion_final = px.bar(situacion_final, x='NOM_REG_RBD_A', y='ASISTENCIA', color='SIT_FIN', barmode='group',
             height=400,width=2000, text_auto=True,title="Situación final según región, teniendo en cuenta promedio de asistencia")
fig_situacion_final.show()
```

## Análisis del gráfico

En el gráfico que antecede se puede ver que en gran parte de las regiones, ante una situacion de desaprobado del alumno, nos encontramos con una asistencia a clases mucho menor que la que se corresponde en promedio con alumnos que han promovido.

# Análisis de la distribución de los alumnos según la jornada en la que asiste



```python
#Se busca conocer la distribución de los alumnos segun la jornada donde asiste
criterio_jornada_establecimiento={'COD_JOR':'count'}
jornada_establecimiento=rendimiento.groupby(['NOM_REG_RBD_A','COD_JOR']).agg(criterio_jornada_establecimiento)
jornada_establecimiento.columns=['CANTIDAD']
jornada_establecimiento=jornada_establecimiento.reset_index()
jornada_establecimiento
```

## Códigos de Jornada
| Código de Jornada  |   Jornada  |
|---|---|
| 1   |   Mañana  |
| 2   |   Tarde  |
| 3   |   Mañana y tarde  |
| 4   |   Vespertina / Nocturna  |

```python
plt.figure(figsize=(20,20))
sns.set_theme(style="ticks", font_scale=1.5)#Se precisaba agrandar la fuente
# en la figura se busca el nombre de la region, la cantidad de alumnos y el codigo de jornada
fig_jornada_establecimiento=sns.barplot(data=jornada_establecimiento, x='NOM_REG_RBD_A', y='CANTIDAD', hue='COD_JOR')
fig_jornada_establecimiento.set(xlabel="Regiones",ylabel="Cantidad de alumnos",title="CANTIDAD DE ALUMNOS SEGÚN TURNO SELECCIONADO")
hands, labs = fig_jornada_establecimiento.get_legend_handles_labels()
plt.legend(handles=hands, labels=['Mañana', 'Tarde', 'Mañana y tarde', 'Vespertina/Nocturna'])
plt.show()
```

# Análisis del gráfico

Es para destacar la predominancia en todos las regiones de la elección, por parte de los alumnos, del turno Mañana y tarde (código 3), seguido por el turno Mañana.

# Análisis de la distribución de alumnos según el código de dependencia en la cual asiste
## Códigos de Dependencia
| Código de Dependencia  |   Dependencia  |
|---|---|
| 1   |   Corporación Municipal  |
| 2   |   Municipal DAEM  |
| 3   |   Particular Subvencionado  |
| 4   |   Particular Pagado (o no subvencionado)  |
| 5   |   Corporación de Administración Delegada (DL 3166)  |

## Género alumno

| Código de genero  |   genero  |
|---|---|
| 0   |   Sin información  |
| 1   |   Masculino  |
| 2   |   Femenino  |

```python
codigo_dep_criterio={'GEN_ALU':'count'}
cod_dep=rendimiento.groupby(['COD_DEPE','GEN_ALU']).agg(codigo_dep_criterio)
cod_dep.columns=['CANTIDAD']
cod_dep=cod_dep.reset_index()
dict_codep={1:'Corporación Municipal',2:'	Municipal DAEM', 3:'Particular Subvencionado', 4:'Particular Pagado (o no subvencionado)',5:'Corporación de Administración Delegada (DL 3166)',6:'Servicio Local de Educación'}
cod_dep['COD_DEPE'] = cod_dep.COD_DEPE.replace(dict_codep)
dict_genalu_codep={0:"Sin información",1:'Masculino',2:'Femenino'}
cod_dep['GEN_ALU'] = cod_dep.GEN_ALU.replace(dict_genalu_codep)
cod_dep
```

```python

codigodependencia = cod_dep.pivot("COD_DEPE", "GEN_ALU", 'CANTIDAD')
f, ax = plt.subplots(figsize=(10, 10))
sns.set_theme(style="ticks", font_scale=1)
mapacalorcodigo=sns.heatmap(codigodependencia, annot=True, fmt=".0f", linewidths=.2, ax=ax)
```

## Análisis del gráfico
Del gráfico que antecede se desprende que claramente hay una cantidad muchisimo mayor de alumnos que se encuentran dentro de dependencias particulares subvencionadas. Seguido por las que son de la clase Municipal DAEM.
