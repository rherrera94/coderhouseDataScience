# **Limpieza de los datos**

Para un mejor entendimiento de la información contenida dentro de ciertos campos lo que se hará es cambiar los nombres de inglés al castellano. Tambien dentro de éste apartado se cambiaran los elementos que vienen como nulos a 0 dentro del campo retraso_llegada.

```python
#cambio el tipo de viaje a castellano
dict_tipo={'Business travel':'Viaje de negocios','Personal Travel':'vacaciones'}
aerolinea['tipo_viaje'] = aerolinea.tipo_viaje.replace(dict_tipo)
aerolinea['tipo_viaje'].value_counts()
```

```python
#cambio el nivel de satisfaccion a castellano
dict_satisfaccion={'neutral or dissatisfied':'neutral o no satisfecho','satisfied':'satisfecho'}
aerolinea['satisfaccion'] = aerolinea.satisfaccion.replace(dict_satisfaccion)
aerolinea['satisfaccion'].value_counts()
```

```python
#cambio el genero a castellano
dict_gen={'Male':'Masculino','Female':'Femenino'}
aerolinea['genero'] = aerolinea.genero.replace(dict_gen)
```

```python
#Los elementos nan deben ser limpiados por lo que realizo la limpieza y reviso si despues de ella no quedan mas por realizar, los que son nan los pondre como 0
aerolinea.retraso_llegada.fillna(0, inplace=True)
aerolinea['retraso_llegada'].unique()
```

## Analisis de la distribución según el género del pasajero

Se busca saber la distribución general según el genero pero también se hará una evaluación de la preferencia de los tipos de asiento. En el caso de los asientos lo que necesitaré hacer es agrupar a los pasajeros primero según la clase elegida y despues según el genero.

```python
#Debemos establecer el criterio de que se contaran los pasajeros por género
distribucion_genero_general = (aerolinea.genero.value_counts()
                                    .to_frame('CANTIDAD').reset_index().rename(columns={'index':'genero'}))
#Criterio por clase de asiento
distribucion_genero_asiento= (aerolinea.groupby(['clase','genero']).genero.count()
                               .to_frame('CANTIDAD').reset_index().rename(columns={'index':'genero'}))
```

# Pasaje de variables categoricas a código

Con el fin de poder realizar la matriz de correlación y lograr considerar cuales son los items con los cuales me quedaré para el análisis se hace necesaria la codificación de las variables categóricas. También éste procedimiento me permitirá después poder realizar los modelos.

Primero veo que, en cuanto al genero, hay dos posibles respuestas que son del tipo string (Masculino/Femenino) por ello se generará un codigo 0 para el Masculino y 1 para el Femenino. Mismo procedimiento se realizará con el tipo de viaje, la clase elegida, el tipo de cliente y con la variable target (satisfacción).

```python
dict_gen={'Masculino':0,'Femenino':1}
aerolinea['genero'] = aerolinea.genero.replace(dict_gen)
dict_viaje={'vacaciones':0,'Viaje de negocios':1}
aerolinea['tipo_viaje'] = aerolinea.tipo_viaje.replace(dict_viaje)
dict_clase={'Eco':0,'Eco Plus':1,'Business':2}
aerolinea['clase'] = aerolinea.clase.replace(dict_clase)
dict_tipo={'disloyal Customer':0,'Loyal Customer':1}
aerolinea['tipo_cliente'] = aerolinea.tipo_cliente.replace(dict_tipo)
dict_satisfaccionv2={'neutral o no satisfecho':0,'satisfecho':1}
aerolinea['satisfaccion'] = aerolinea.satisfaccion.replace(dict_satisfaccionv2)
```

# Grafico matriz para ver la relacion entre las variables

Para tener una buena visión de la correlación entre las diferentes variables es que creare un heatmap que muestre éstos valores de manera gráfica y poder valorar cuan positivas o negativas son esas relaciones.

```python
aerolineaaux=aerolinea.drop(['satisfaccion'],axis=1)
mapacoef = plt.cm.viridis
plt.figure(figsize=(40,40))
plt.title('Heatmap Coeficiente de correlación', y=1.05, size=25)
sns.heatmap(aerolineaaux.astype(float).corr(),linewidths=0.1,vmax=1.0, square=True, cmap=mapacoef, linecolor='white', annot=True)
plt.show()
```

## Modelos

```python
#Metricas resultantes de las predicciones
metricas_modelos=pd.DataFrame(columns=['modelo','precision_test','precision_entrenamiento','accuracy_test','accuracy_entrenamiento','recall_test','recall_entrenamiento','f1_test','f1_entrenamiento'])
metricas_kfold=pd.DataFrame(columns=['modelo','precision','accuracy'])
#arrays auxiliares
listado_modelos=['Arbol de decisión','Naive bayes','Regresión logistica','Random Forest','Gradient Boosting classifier','Linear discriminant analysis','XGBOOST','AdaBoostClassifier']
tiempos_entrenamiento=[]

############################################################################
############################Funciones Auxiliares############################
############################################################################


def matriz_confusion(ytesteo,yprediccion_testeo,modelo,label):
  confusion=ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(ytesteo, yprediccion_testeo),display_labels=label)
  fig, ax = plt.subplots(figsize=(10,10))
  confusion.plot(cmap="OrRd",ax=ax)
  ax.set_title("Matriz de confusión "+modelo+" (test): ")
  ax.set_xlabel ("Predicción")
  ax.set_ylabel ("Satisfacción real")
  plt.show()

def hist_probabilidad (proba0,proba1,ytesteo,modelo):
  plt.figure(figsize=(13,13))
  plt.hist(proba0,bins=50,label='Probabilidad de 1')
  plt.hist(proba1,bins=50,label='Probabilidad de 0',color='y')
  plt.hist(ytesteo,bins=50,label='Respuesta correcta',alpha=0.7,color='r')
  plt.xlabel('Probabilidad de predecida vs respuesta correcta')
  plt.ylabel('Cantidad de registros')
  plt.title(modelo)
  plt.legend(fontsize=15)
  plt.tick_params(axis='both',labelsize=25,pad=5)
  plt.show()

def importancia_feature(modelo,xentrenamiento):
  sorted_idx = modelo.feature_importances_.argsort()
  plt.figure(figsize=(13,13))
  plt.barh(xentrenamiento.columns[sorted_idx], modelo.feature_importances_[sorted_idx])
  plt.xlabel("Nivel de importancia")
  plt.legend(fontsize=15)
  plt.show()

def shap_dibujar(modelo,xentrenamiento):
  explainer=shap.explainers.Linear(modelo,xentrenamiento)
  shap_values = explainer(xentrenamiento)
  shap.summary_plot(shap_values, xentrenamiento, plot_type="bar")

def curva_roc(modelo,false_positive_rate1, true_positive_rate1):
  plt.subplots(1, figsize=(8,6))
  plt.title('Receiver Operating Characteristic - '+modelo)
  plt.plot(false_positive_rate1, true_positive_rate1)
  plt.plot([0, 1], ls="--")
  plt.plot([0, 0], [1, 0] , c=".7"), plt.plot([1, 1] , c=".7")
  plt.ylabel('True Positive Rate')
  plt.xlabel('False Positive Rate')
  plt.show()
```

# Modelos

```python
i=0
for models, lista in zip(modelos,listado_modelos):

  #start y end los utilizo para contabilidar el tiempo de entrenamiento
  start=time.time()
  #entrena los modelos descriptos en el listado de modelos que utilizare
  modelo = models.fit(xentrenamiento,yentrenamiento)
  end=time.time()
  
  #Predicciones
  yprediccion_entrenamiento=modelo.predict(xentrenamiento)
  yprediccion_testeo=modelo.predict(xtesteo)
  
  #predigo la probabilidad en cada uno de los casos
  proba0=modelo.predict_proba(xtesteo)[:,0]
  proba1=modelo.predict_proba(xtesteo)[:,1]

  #Guardo los resultados para despues poder realizar analisis
  tiempos_entrenamiento.append(end-start) 
  metricas_modelos = metricas_modelos.append({
        'modelo':lista,
        'precision_test':precision_score(ytesteo, yprediccion_testeo) ,
        'precision_entrenamiento':precision_score(yentrenamiento, yprediccion_entrenamiento),
        'accuracy_test':accuracy_score(ytesteo, yprediccion_testeo),
        'accuracy_entrenamiento':accuracy_score(yentrenamiento, yprediccion_entrenamiento),
        'recall_test':recall_score(ytesteo, yprediccion_testeo),
        'recall_entrenamiento':recall_score(yentrenamiento, yprediccion_entrenamiento),
        'f1_test':f1_score(ytesteo, yprediccion_testeo),
        'f1_entrenamiento':f1_score(yentrenamiento, yprediccion_entrenamiento)
        }, ignore_index=True)
  
  #Muestro los reportes de cada uno de los modelos
  display(Markdown('## <ins>Classification report de %s (test):</ins>'%(lista)))
  print(classification_report(ytesteo,yprediccion_testeo,digits=7))

  display(Markdown('## <ins>Explicación de las metricas de Classification report de %s (test):</ins>'%(lista)))
  print('(nota aclaratoria: neutral o no satisfecho:0,satisfecho:1 )')
  
  print('- En cuanto a la precisión del modelo (is the ability of the classifier not to label as positive a sample that is negative), éste acerto en un porcentaje del %.5f  porciento que el pasajero estaba satisfecho,'%(precision_score(ytesteo, yprediccion_testeo)*100))
  print('mientras que en cuanto al porcentaje de precision sobre los pasajeros que no estaban satisfechos fue del %.5f porciento' %(precision_score(ytesteo, yprediccion_testeo,pos_label=0)*100)) 
  print('- El accuracy (the set of labels predicted for a sample must exactly match the corresponding set of labels in y_true) del modelo nos dice que éste predijo en un porcentaje del %.5f  la satisfacción de manera correcta' %(accuracy_score(ytesteo, yprediccion_testeo)*100))
  print('- Por definición la función del recall es: intuitively the ability of the classifier to find all the positive samples')
  print('en este caso el modelo tuvo la abilidad de encontrar sobre el total de los pasajeros satisfechos el %.5f porciento de ellos' %(recall_score(ytesteo, yprediccion_testeo)*100))

  print("------------------------------------------------------------------------------------------")
  display(Markdown('## <ins>Classification report de %s (entrenamiento):</ins>'%(lista)))
  print(classification_report(yentrenamiento,yprediccion_entrenamiento,digits=7))
  print("-------------------------------------------------------------------------------------------------------")

  #Matriz de confusión
  display(Markdown('## <ins>Matriz de confusión %s:</ins>'%(listado_modelos[i])))
  matriz_confusion(ytesteo,yprediccion_testeo,lista,['neutral o no satisfecho','satisfecho'])
  print("-------------------------------------------------------------------------------------------------------")

  # Feature importance
  if(lista!='Naive bayes' and lista!='Regresión logistica' and lista!='Linear discriminant analysis'):
   display(Markdown('## <ins>Feature Importance %s:</ins>'%(lista)))
   importancia_feature(modelo,xentrenamiento)
  
  #shap
  if (lista=='Regresión logistica' or lista=='Linear discriminant analysis'):
    display(Markdown('## <ins> Feature Importance Shap %s:</ins>'%(lista)))
    shap_dibujar(modelo,xentrenamiento)

  # Comienza el histograma correspondiente a la probabilidad predecida
  display(Markdown('## <ins>Histograma de probabilidad %s:</ins>'%(lista)))
  hist_probabilidad (proba0,proba1,ytesteo,lista)

  #Curva ROC
  # Calculo de tasas 
  display(Markdown('## <ins> Curva ROC %s:</ins>'%(lista)))
  false_positive_rate1, true_positive_rate1, threshold1 = roc_curve(ytesteo, proba1)
  print('roc_auc_score: ', roc_auc_score(ytesteo, proba1))
  curva_roc(lista,false_positive_rate1, true_positive_rate1)

  #Cross-validation
  display(Markdown('## <ins> Cross-validation %s:</ins>'%(lista)))
  cv_results = cross_val_score(models, x, y, cv=10, scoring='accuracy')
  print('Accuracy: %.3f (%.3f)' %(np.mean(cv_results), np.std(cv_results)))
  display(Markdown('## <ins> Stratified K-Fold %s:</ins>'%(lista)))
  list_ac=[]
  list_pre=[]
  for h in range(0,skf.get_n_splits(x, y)):
    # Calculo de indices
    train, test = list(skf.split(x, y))[h]
    # Separar train y test
    X_train_n, y_train_n, X_test_n, y_test_n= x.iloc[train], y.iloc[train], x.iloc[test],y.iloc[test]
    # Ajustar el modelo
    models.fit(X_train_n, y_train_n)
    # Predicciones
    pred= models.predict(X_test_n)
    # Calculo del error
    accuracy= accuracy_score(y_true=y_test_n, y_pred= pred)
    precision= precision_score(y_true=y_test_n, y_pred= pred)
    print('Iteracion: ',h+1, ' Accuracy:', accuracy)
    print('Iteracion: ',h+1, ' Precision:', precision)
    list_ac.append(accuracy)
    list_pre.append(precision)
  # Promedio de accuracy
  print('Accuracy promedio:', np.mean(list_ac))
  print('Precision promedio:', np.mean(list_pre))
  metricas_kfold = metricas_kfold.append({
        'modelo':lista,
        'precision':np.mean(list_pre) ,
        'accuracy':np.mean(list_ac)
        }, ignore_index=True)
  print("------------------------------------------------------------------------------------------------------------")
  print("------------------------------------------------------------------------------------------------------------")
  i=i+1

display(Markdown('## <ins>Métricas de los modelos</ins>'))
metricas_modelos.set_index('modelo', inplace=True)
display(metricas_modelos)
display(Markdown('## <ins>Métricas de los modelos K-fold</ins>'))
metricas_kfold.set_index('modelo', inplace=True)
display(metricas_kfold)
```

## Conclusion cross validation con K-fold

Entiendo que por las metricas promedio que da, que no son alejadas a las que dio por el metodo convencional, y el coste computacional que significa utilizar este tipo de procedimiento no es conveniente utilizar el cross validation sino que conviene el modelo convencional.

Las razones para llegar a esta conclusión son las siguientes:

1) el modelo se tendra que correr varias veces mas (tantas veces como cantidad de folds que elija) esto significa que si por el metodo convencional lo corro 1 vez y en k-fold defino k=10 hara 10 veces mas de trabajo que en el convencional.

2) las metricas que me dio en K-fold no son muy lejanas a las que dio en el metodo convencional.

Es de aclarar que en el caso del modelo convencional se usa una semilla por lo que siempre va a dar los mismos valores de puntaje, ahora en el modelo K-fold no se utiliza esta semilla por lo que en estos casos lo que sucedera es que siempre se utilizaran sets diferentes y por lo tanto en cada vuelta dara diferente (a diferencia del metodo tradicional, que le puse semilla) y por ello tambien que en el metodo k-fold se utiliza un promedio de los puntajes. En mi caso prefiero seleccionar el metodo tradicional, como dije antes por poseer un menor coste computacional y al no ser una diferencia muy grande de puntaje hace que no justifique el coste computacional de utilizar el metodo K-fold.
